```python
import seaborn as sns
import pandas as pd
import os

results_path = os.path.join("results", "models_results.csv")

df = pd.read_csv(results_path)
df.head()
```

```python
print("Best Acurracys :")
df.sort_values(by=["accuracy"], ascending=[False]).head(10)
```

```python
print("Best Acurracys for each Model :")

df.loc[df.groupby(["algorithm"])["accuracy"].idxmax()].set_index("algorithm")\
                                                      .sort_values(by=["accuracy"], ascending=[False])
```

From here we can see that the majority of models prefered the pre_processed dataset

```python
#Create a list of all the models that benefit from data scaling
models_to_scale = ['KNN', 'SVC', 'Bernoulli_NB', 'Multinomial_NB', 'Gaussain_NB']

#Create a list of all the models that DOESNT benefit from data scaling
models_without_scale = ['AdaBoost', 'CatBoost', 'Decision_Tree_Classifier', 'Extra_Tree_Classifier', 'Extra_Trees_Classifier', 'GradientBoost', 'LGBM', 'Random_Forest_Classifier']
```

```python
df.loc[df["algorithm"].isin(models_to_scale)].groupby(["algorithm", "vectorizer"], as_index = True)\
                                             .mean(numeric_only = True)\
                                             .sort_values(by=["algorithm","accuracy"], ascending=[True, False])\
                                             [["accuracy", "train_time", "predict_time"]]
```

```python
df.loc[df["algorithm"].isin(models_without_scale)].groupby(["algorithm", "vectorizer"], as_index = True)\
                                                  .mean(numeric_only = True)\
                                                  .sort_values(by=["algorithm","accuracy"], ascending=[True, False])\
                                                  [["accuracy", "train_time", "predict_time"]]
```

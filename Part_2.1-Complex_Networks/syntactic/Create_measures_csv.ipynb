```python
import networkx as nx
from tqdm.notebook import tqdm
import os
import statistics
import pandas as pd

import sys
sys.path.append( '.' )
import importlib

import measures
importlib.reload(measures)
```

```python
def add_row_to_df(df, i, label, measures, news_id):
    
    df.loc[i, 'news_id'] = news_id
    df.loc[i, 'betweenness'] = measures['betweenness']
    df.loc[i, 'closeness'] = measures['closeness']
    df.loc[i, 'eigenvector'] = measures['eigenvector']
    df.loc[i, 'katz'] = measures['katz']
    df.loc[i, 'pagerank'] = measures['pagerank']
    df.loc[i, 'hubs'] = measures['hubs']
    df.loc[i, 'authorities'] = measures['authorities']
    df.loc[i, 'density'] = measures['density']
    df.loc[i, 'label'] = label

    return df
```

```python
df = pd.DataFrame(columns=['news_id', 'betweenness', 'closeness', 'eigenvector', 'katz', 'pagerank', 'hubs', 'authorities', 'density', 'label'])

i = 0
for file in tqdm(os.listdir('data/udpipe_tuples_with_stopwords/' + 'fake')):

    fake_file_path = 'data/udpipe_tuples_with_stopwords/' + 'fake' + '/' + file
    true_file_path = 'data/udpipe_tuples_with_stopwords/' + 'true' + '/' + file

    G_fake = measures.generate_graph_udpipe(fake_file_path)
    G_true = measures.generate_graph_udpipe(true_file_path)

    measures_fake = measures.get_measures_from_graph(G_fake)
    measures_true = measures.get_measures_from_graph(G_true)

    for key, value in list(measures_fake.items()):
        if  isinstance(value, dict):
            measures_fake[key] = statistics.mean(value.values())

    for key, value in list(measures_true.items()):
        if  isinstance(value, dict):
            measures_true[key] = statistics.mean(value.values())
    
    news_id = file.split('.')[0]

    add_row_to_df(df, i, "fake", measures_fake, news_id)
    add_row_to_df(df, i+1, "true", measures_true, news_id)

    i += 2
```

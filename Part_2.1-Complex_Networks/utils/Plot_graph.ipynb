```python
import networkx as nx

G = nx.DiGraph() # Grafo direcionado
```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt

file_path = os.path.join('..','..', 'data', 'csvs', 'pre_normalized.csv')
df = pd.read_csv(file_path)
df.head(3)
```

```python
test_sentence = df["text"].values[1]
test_sentence_words = test_sentence.split()

G.add_nodes_from(test_sentence_words)

for i in range(len(test_sentence_words) - 1):
    G.add_edge(test_sentence_words[i], test_sentence_words[i + 1])
```

```python
nx.draw_circular(G, with_labels=True, node_size=15, font_size=7, edge_color='gray')
plt.savefig('../results/adjacency_graph.png', dpi=500)
```
